```python
# 03_resume_checker.ipynb

import os
import re
import pandas as pd
from collections import Counter

# Optional: If parsing PDF resumes
# !pip install PyPDF2 spacy
# import spacy
# nlp = spacy.load("en_core_web_sm")

# --------------------------------------------------------
# Step 1: Load resume text
# For demo, you can paste plain text OR read from a PDF file
resume_path = r"D:\project hack\data\raw\last resume.pdf"  # change to .pdf if needed

def extract_text_from_file(file_path):
    if file_path.endswith(".txt"):
        with open(file_path, "r", encoding="utf-8") as f:
            return f.read()
    elif file_path.endswith(".pdf"):
        from PyPDF2 import PdfReader
        reader = PdfReader(file_path)
        text = ""
        for page in reader.pages:
            text += page.extract_text() + "\n"
        return text
    else:
        raise ValueError("Unsupported resume format. Use .txt or .pdf")

resume_text = extract_text_from_file(resume_path)
print("✅ Resume loaded")
print(resume_text[:500])  # preview first 500 chars

# --------------------------------------------------------
# Step 2: Skills extraction
# Define a basic skills library (extend with your dataset’s skills)
skills_master = [
    "python", "java", "c++", "c", "c#", "sql", "mysql", "mongodb", "html", "css", "javascript",
    "machine learning", "deep learning", "nlp", "pandas", "numpy", "tensorflow", "keras",
    "flask", "django", "react", "angular", "git", "docker", "cloud", "aws", "azure", "linux"
]

def extract_skills(text, skills_list):
    text = text.lower()
    found = []
    for skill in skills_list:
        if re.search(r"\b" + re.escape(skill.lower()) + r"\b", text):
            found.append(skill.lower())
    return sorted(set(found))

resume_skills = extract_skills(resume_text, skills_master)
print("Extracted Skills:", resume_skills)

# --------------------------------------------------------
# Step 3: Education & Experience extraction (basic regex)
def extract_education(text):
    degrees = ["btech", "mtech", "bachelor", "master", "phd", "b.sc", "m.sc", "mba"]
    found = []
    for d in degrees:
        if re.search(r"\b" + d + r"\b", text.lower()):
            found.append(d)
    return list(set(found))

def extract_experience(text):
    exp_match = re.findall(r"(\d+)\s+(?:years|yrs)\s+experience", text.lower())
    return int(exp_match[0]) if exp_match else 0

education = extract_education(resume_text)
experience_years = extract_experience(resume_text)

print("Education:", education)
print("Experience (years):", experience_years)

# --------------------------------------------------------
# Step 4: Build User Profile
user_profile = {
    "skills": resume_skills,
    "education": education,
    "experience_years": experience_years,
    "location_preference": ["delhi", "remote"],  # fallback (could parse address from resume)
    "min_stipend": 5000,
    "duration_preference": None
}

print("\n✅ User Profile Generated:")
print(user_profile)

# --------------------------------------------------------
# Step 5: Save User Profile for next notebook
PROCESSED_DIR = r"D:\project hack\data\processed"
os.makedirs(PROCESSED_DIR, exist_ok=True)

profile_path = os.path.join(PROCESSED_DIR, "user_profile.json")
import json
with open(profile_path, "w", encoding="utf-8") as f:
    json.dump(user_profile, f, indent=4)

print(f"\n✅ User Profile saved at: {profile_path}")
```
