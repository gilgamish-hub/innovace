```python
# 02_user_profile_filtering.ipynb

import os
import pandas as pd

# Paths
PROCESSED_DIR = r"D:\project hack\data\processed"
CLEAN_DATA_PATH = os.path.join(PROCESSED_DIR, "internship_data_clean.csv")

# Load cleaned dataset
df = pd.read_csv(CLEAN_DATA_PATH)
print("✅ Cleaned dataset loaded")
print("Shape:", df.shape)
display(df.head(5))

# -------------------------------------------------------
# Step 1: Define User Profile
# (In real app, this will come from user input or resume parsing)
user_profile = {
    "skills": ["python", "machine learning", "sql"],
    "location_preference": ["delhi", "remote"],
    "min_stipend": 5000,
    "duration_preference": None  # e.g. "6 months", None means no filter
}

print("User Profile:", user_profile)

# -------------------------------------------------------
# Step 2: Basic filters

filtered_df = df.copy()

# Location filter
if user_profile["location_preference"]:
    filtered_df = filtered_df[
        filtered_df["location"].str.lower().isin([loc.lower() for loc in user_profile["location_preference"]])
    ]

# Stipend filter
if "stipend_numeric" in filtered_df.columns:
    filtered_df = filtered_df[filtered_df["stipend_numeric"] >= user_profile["min_stipend"]]

# Duration filter
if user_profile["duration_preference"] and "duration" in filtered_df.columns:
    filtered_df = filtered_df[filtered_df["duration"].str.contains(user_profile["duration_preference"], case=False, na=False)]

print("After applying filters:", filtered_df.shape)

# -------------------------------------------------------
# Step 3: Skills matching
if "skills" in filtered_df.columns:
    def skill_match(row_skills, user_skills):
        row_skills = str(row_skills).lower().split(",")
        return len(set(row_skills) & set([s.lower() for s in user_skills]))
    
    filtered_df["skill_match_count"] = filtered_df["skills"].apply(
        lambda x: skill_match(x, user_profile["skills"])
    )
    
    # Keep only internships with at least 1 skill match
    filtered_df = filtered_df[filtered_df["skill_match_count"] > 0]

print("After skill matching:", filtered_df.shape)

# -------------------------------------------------------
# Step 4: Rank by skill match + stipend
if "stipend_numeric" in filtered_df.columns:
    filtered_df = filtered_df.sort_values(
        by=["skill_match_count", "stipend_numeric"], ascending=[False, False]
    )

# Save filtered results
filtered_path = os.path.join(PROCESSED_DIR, "internship_data_filtered.csv")
filtered_df.to_csv(filtered_path, index=False, encoding="utf-8")
print(f"✅ Filtered internships saved at: {filtered_path}")

# Show top 10 recommendations (before full recommendation engine)
display(filtered_df.head(10))
```
