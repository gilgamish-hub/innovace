```python
# 01_dataset_setup_eda.ipynb

import os
import pandas as pd

# Paths
RAW_DATA_PATH = r"D:\project hack\data\raw\internship_data.csv"
PROCESSED_DIR = r"D:\project hack\data\processed"
os.makedirs(PROCESSED_DIR, exist_ok=True)

# Load dataset
df = pd.read_csv(RAW_DATA_PATH)
print("✅ Dataset loaded")
print("Shape:", df.shape)

# Quick overview
print("\nColumns:", df.columns.tolist())
print("\nData Types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum())

# Preview
display(df.head(10))

# -----------------------------------------
# Step 1: Drop duplicates
df = df.drop_duplicates()
print("After removing duplicates:", df.shape)

# Step 2: Handle missing values (basic example: fill NA with empty string for text cols)
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].fillna("")

# Step 3: Normalize column names
df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]

# Step 4: Basic stipend cleaning (₹ → numeric, monthly)
if "stipend" in df.columns:
    import re
    def parse_stipend(s):
        if pd.isna(s): return None
        s = str(s).replace("₹", "").replace(",", "").strip().lower()
        match = re.findall(r"\d+", s)
        return int(match[0]) if match else None
    df["stipend_numeric"] = df["stipend"].apply(parse_stipend)

# Step 5: Save cleaned dataset
processed_path = os.path.join(PROCESSED_DIR, "internship_data_clean.csv")
df.to_csv(processed_path, index=False, encoding="utf-8")
print(f"✅ Cleaned dataset saved at: {processed_path}")

# Summary stats
print("\nSummary:")
print(df.describe(include="all").transpose().head(15))
```
